```python
import numpy as np
import pandas as pd
```

```python
# Import comma-separated variable file
#pollutant one is NO2

pollutant_one = pd.read_csv('../data/NO2_2010_2020.csv', sep=';')
pollutant_one.drop(pollutant_one.iloc[:,:1], inplace = True, axis = 1) 
pollutant_one.columns = 'NO2_' + pollutant_one.columns

# The data is for all the regions in NL. Now we need to filter Amsterdam data among all the NL Neighbourhoods.

amsterdam_pollutant_one = pollutant_one[pollutant_one['NO2_name'].str.contains('Amsterdam', na=False)]
amsterdam_pollutant_one.head(5)
```

```python
#pollutant two is PM2.5

pollutant_two = pd.read_csv('../data/PM2.5_2010_2020.csv', sep=';')
pollutant_two.drop(pollutant_two.iloc[:,:1], inplace = True, axis = 1) 
pollutant_two.columns = 'PM2.5_' + pollutant_two.columns

# The data is for all the regions in NL. Now we need to filter Amsterdam data among all the NL Neighbourhoods.

amsterdam_pollutant_two = pollutant_two[pollutant_two['PM2.5_name'].str.contains('Amsterdam', na=False)]
amsterdam_pollutant_two.head()
```

```python
#pollutant three is PM10

pollutant_three = pd.read_csv('../data/PM10_2010_2020.csv', sep=';')
pollutant_three.drop(pollutant_three.iloc[:,:1], inplace = True, axis = 1) 
pollutant_three.columns = 'PM10_' + pollutant_three.columns

# The data is for all the regions in NL. Now we need to filter Amsterdam data among all the NL Neighbourhoods.

amsterdam_pollutant_three = pollutant_three[pollutant_three['PM10_name'].str.contains('Amsterdam', na=False)]
amsterdam_pollutant_three.head()
```

```python
# Now we need to identify the neighbourhoods that are common in the 3 pollutant tables above. 
# In order to do that we first merger the first 2 tables and then we merge the resulatant table and the third table

merged_inner = pd.merge(left=amsterdam_pollutant_one, right=amsterdam_pollutant_two, left_on='NO2_name', right_on='PM2.5_name')
merged_inner = pd.merge(left=merged_inner, right=amsterdam_pollutant_three, left_on='NO2_name', right_on='PM10_name')
merged_inner
```
